```python
n=30 #year before retirement
t=10 #year after retirement
g=0.02 #growing rate
r=0.03 #saving return rate
c=500000 #current annual consumpment
c1=c*(1+g)**n #退休第一年所需消費
PV=c1*(1-((1+g)/(1+r))**t)/(r-g) #退休共需多少消費
s=PV*(r-g)/(1-((1+g)/(1+r))**n) #現在每年應儲蓄金額
```

```python
print(c1)
```

```python
print(PV)
```

```python
print(s)
```
