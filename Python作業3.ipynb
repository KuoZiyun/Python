```python
def pv(rate, cashflows):
    npv = 0
    for t, cf in enumerate(cashflows):
        npv += cf / (1 + rate)**t
    return npv
```

```python
cashflows = [-100,40,40,40]
low = -0.9999
high = 10
middle = (high + low) / 2
tolerance = 0.00001

while abs(pv(middle, cashflows)) > tolerance: #當pv(middle)的絕對值>0.00001，進入迴圈

    # 如果pv(low)*pv(middle)>0 and pv(low)>0, 表示同號, 設middle為新的low
    if pv(low, cashflows) * pv(middle, cashflows) > 0 and pv(low, cashflows) > 0:
        low = middle

    # 如果pv(low)*pv(middle)>0 and pv(low)<0, 表示同號, 設low為新的high
    elif pv(low, cashflows) * pv(middle, cashflows) > 0 and pv(low, cashflows) < 0:
        high = low

    # 如果pv(low)*pv(middle)<0, 表示異號, 設middle為新的higt
    elif pv(low, cashflows) * pv(middle, cashflows) < 0:
        high = middle
    middle = (high + low) / 2
    print(middle)
```
