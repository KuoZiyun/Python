```python
a=input('請輸入目前景氣對策燈號分數')

try:
  if int(a)>=38:
    print('紅燈，表示景氣過熱，應採取緊縮措施。建議賣出')

  elif int(a)>=32 and int(a)<=38:
    print('黃紅燈，表示景氣活絡，但短期內可能轉熱，不宜再實施刺激政策。建議維持現狀')

  elif int(a)>=21 and int(a)<=31:
    print('綠燈，代表景氣穩定，是理想的經濟狀態，可採穩定促進成長的措施。建議維持現狀')

  elif int(a)>=17 and int(a)<=22:
    print('黃藍燈，表示景氣欠佳，介於穩定與低迷之間，需要密切觀察是否會進一步惡化。建議維持現狀')

  elif int(a)>=9 and int(a)<=16:
    print('藍燈，表示景氣低迷。建議買進')

except:
  print('請輸入有效數字')
```
