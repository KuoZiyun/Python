```python
# 3. while loop to print the list
x = ['a', 'b', 'c', 'd']
i = 0
while i < len(x):
    print(x[i])
    i += 1
```

```python
# 4. for loop to print the dictionary
x = {1: 'a', 2: 'b', 3: 'c', 4: 'd'}
for x, y in x.items():
    print(x, y)
```

```python
# 6. sort() vs sorted()
# sort()
a = [3, 1, 4, 1, 5, 9, 2, 6]
a.sort()
print(a)

# sorted()
a = [3, 1, 4, 1, 5, 9, 2, 6]
sorted_a = sorted(a)
print(sorted_a)
print(a)
```

```python
# 7. combining list
a = [i**2 for i in range(5) if i%2 == 0]
a = []
for i in range(5):
    if i % 2 == 0:
        a.append(i**2)

print(a)
```

```python
# 10.mutable vs immutable

# mutable objects: include 'list','dict','set'. A mutable object, such as a list, will always refer to the same id.
x1 = ["iphone", "ipad", "macbook"]
x1[2] = 'airpods'
print (x1)
```

```python
# 10.mutable vs immitable

#immutable objects:include 'int','float','string','tuple'.Immutable objects are more efficient in reading and writing data from/into a computer's memory, but are costly when changing the value.
a1 = 'abc'
a2 = a1
print(id(a1))
print(id(a2))
```
