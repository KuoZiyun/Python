```python
# (1)日期格式修正+排序

import pandas as pd

url = "/content/drive/MyDrive/Python經濟/csv.csv"
df = pd.read_csv(url)

# 刪除'備註'列
df = df.drop(columns=['備註'])

# 轉換為西元年
df['Gregorian_Year'] = df['年度'] + 1911

# 西元年-月份-日期
df['日期'] = pd.to_datetime(df['Gregorian_Year'].astype(int).astype(str) + '-' + df['月份'].astype(int).astype(str) + '-01')
df.set_index('日期', inplace=True)
df.drop(columns=['年度', '月份', 'Gregorian_Year'], inplace=True)

display(df.head())

print(df.dtypes)
```

```python
# (2)所有numeric欄位轉成float

import pandas as pd

# 交易列
trade_columns = [
    '出口總值(新臺幣千元)',
    '出口(新臺幣千元)',
    '復出口(新臺幣千元)',
    '進口總值(新臺幣千元)',
    '進口(新臺幣千元)',
    '復進口(新臺幣千元)',
    '出入超(新臺幣千元)'
]

# 將數字欄位轉換為float
for col in trade_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('float64')

print(df.dtypes)
```

```python
# (3)建立至少兩個新變數

import pandas as pd

# 計算'總貿易額(新臺幣千元)'
df['總貿易額(新臺幣千元)'] = df['出口總值(新臺幣千元)'] + df['進口總值(新臺幣千元)']

# 計算'淨出口(新臺幣千元)'
df['淨出口(新臺幣千元)'] = df['出口總值(新臺幣千元)'] - df['進口總值(新臺幣千元)']

print(df.dtypes)

display(df.head())
```

```python
# (4)處理缺漏值

import pandas as pd

# 檢查缺失值的交易列
trade_columns_to_check = [
    '出口總值(新臺幣千元)',
    '出口(新臺幣千元)',
    '復出口(新臺幣千元)',
    '進口總值(新臺幣千元)',
    '進口(新臺幣千元)',
    '復進口(新臺幣千元)',
    '出入超(新臺幣千元)',
    '總貿易額(新臺幣千元)',
    '淨出口(新臺幣千元)'
]

# 計算並列印缺失值數量
print(df[trade_columns_to_check].isna().sum())

initial_rows = df.shape[0]

df.dropna(subset=trade_columns_to_check, inplace=True)

# 列印剩餘行數
rows_after_dropping_na = df.shape[0]

display(df.head())
```

```python
# (5)簡單敘述統計（平均數、變異數、25分位數、75分位數、樣本數）

import numpy as np

# 交易列
trade_columns_for_stats = [
    '出口總值(新臺幣千元)',
    '出口(新臺幣千元)',
    '復出口(新臺幣千元)',
    '進口總值(新臺幣千元)',
    '進口(新臺幣千元)',
    '復進口(新臺幣千元)',
    '出入超(新臺幣千元)',
    '總貿易額(新臺幣千元)',
    '淨出口(新臺幣千元)'
]

# 統計資料
statistics = {}

# 計算資料
for col in trade_columns_for_stats:
    statistics[col] = {
        'mean': df[col].mean(),
        'variance': df[col].var(),
        '25th_percentile': df[col].quantile(0.25),
        '75th_percentile': df[col].quantile(0.75),
        'sample_size': df[col].count()
    }

for col, stats in statistics.items():
    print(f"\n--- {col} ---")
    for stat_name, value in stats.items():
        print(f"{stat_name}: {value:,.2f}")
```

```python
# (5)簡單 groupby 的敘述統計（平均數跟變異數）

import pandas as pd

# 抓取年分
df['Year'] = df.index.year

# 按年份分組並計算平均數跟變異數
yearly_trade_stats = df.groupby('Year')[['出口總值(新臺幣千元)', '進口總值(新臺幣千元)']].agg(['mean', 'var'])

display(yearly_trade_stats)
```
